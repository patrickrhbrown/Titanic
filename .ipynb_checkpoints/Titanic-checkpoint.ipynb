# Predicting Survival on the Titanic

The sinking of the RMS Titanic is one of the most infamous shipwrecks in history.  On April 15, 1912, during her maiden voyage, the Titanic sank after colliding with an iceberg, killing 1502 out of 2224 passengers and crew. This sensational tragedy shocked the international community and led to better safety regulations for ships.

One of the reasons that the shipwreck led to such loss of life was that there were not enough lifeboats for the passengers and crew. Although there was some element of luck involved in surviving the sinking, some groups of people were more likely to survive than others, such as women, children, and the upper-class.

In this challenge, taken from Kaggle, we are to complete the analysis of what sorts of people were likely to survive. In particular, we ask you to apply the tools of machine learning to predict which passengers survived the tragedy.

### Data Dictionary

Contained within the dataset are the following features:
    
* PassengerID: a unique identifier integer for each passenger
* Survived: a classifier that denotes whether a passenger survived or died (0 for died, 1 for survived) so in effect a bool
* Pclass: a field denoting the travel class of the passenger as an integer 1, 2 or 3
* Name: the name of the passenger as a string
* Sex: the gender of the passenger (male, female as a string)
* Age: the age of the passenger, as a float
* Sibsp: how many siblings or partners the passenger had onboard as an int
* Parch: number of children or parents onboard as an int
* Ticket: the ticket number of the passenger
* Fare: how much the passenger paid for their ticket (our assumption is from reading this column in conjunction with Pclass column is that 1 is the top Pclass and 3 is the lowest)
* Cabin: the number of the passenger's cabin as a string
* Embarked: the departure point for the passenger (Cherbourg (C), Queenstown (Q) or Southampton (S))

We can verify the dtypes as follows:

```python
Titanic_train_df.info()
```

Clearly we have some missing values which will require tidying up in the case of Age and Cabin.

```python
sns.countplot(x = 'Survived', data = Titanic_train_df)
plt.title("Count Plot of Survival.  {} per cent survived".format(Titanic_train_df['Survived'].round(decimals=0).mean() % 10 * 100))
```

```python
Titanic_train_df.describe()
```

Just under 40 per cent of passengers survived in the training set. A countplot shows that essentially our base model is 'predict survive' regardless of other features, with an accuracy of approx 38%.

How can we improve our model? It's cheesey but perhaps we should take a think about the popular history and see if we can obtain some insights.  First, 'women and children first' is a cliche of disaster films and popular culture, but perhaps that's what happened here, and we know that society was much more explicitly stratified at that point in time.   Let's  group the data by Pclass and see what the mean survival rate was.

```python
Titanic_train_df.groupby('Pclass')['Survived'].mean()
```

Survival clearly improves as you graduate up the passenger class scale with first class passengers showing a survival rate of 62% compared with < 25% for 3rd class passengers.  What about gender?

```python
sns.barplot(x = 'Sex', y = 'Survived', hue = "Pclass", data = Titanic_train_df)
plt.title('Barchart of Survival by Gender and Passenger Class')
```

```python
gender_and_class = Titanic_train_df.groupby(['Pclass','Sex'])['Survived'].mean()
gender_and_class
```
